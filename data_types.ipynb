```python
# int: store whole num
age = int(input("Age ?"))
print(age)
type(age)

# can int store decimal values? 
# No, int store only whole num(+,- or 0), use float to use decimal.
```

```python
# float: store decimal num
temp_in_c = float(input("Enter Temp outside in Delhi"))
print(temp_in_c)
type(temp_in_c)
```

```python
# complex: store imaginary num/  a complex num that has real part and imginary part, it's written as a+bj.
num = 3+4j
print(num)
type(num)

#Ex 2
a = 5 + 2j
b = 2 + 4j
print(a + b)
print(a - b)
print(a * b)

# acces real and imaginary  part, return float val
print(a.real)
print(a.imag)

#create complex val
a1 = complex(4,6)
print(a1)
```

```python
# Boolean: store only 2 values - True & False
is_logged = True
print(is_logged)
type(is_logged)
```

```python
# string: store text
name = input("Enter your name")
print(name)
type(name)
```

```python
# List: Ordered, mutable collection, dublicate allowed.
num1 = [10,20,30,40]
list1 = [10,"Ai",100,12.5]
print(num1)
print(list1)
print(type(num1))
print(type(list1))
```

```python
# Tuple: Orderd but immutable collection.
num = (10,20,30,40,50)
print(num)
type(num)
```

```python
# Set: Unorderd collection, No dublicate value.
num = {10,20,30,40,50}
print(num)
type(num)
```

```python
# Dictionary: store data in key : value
emp = {
    "name":"Ahishek",
    "exp":5,
    "city":"Dwaraka, New Delhi"
}
print(emp)
type(emp)
```
