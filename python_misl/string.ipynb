## Escape Char:
#### that start with a backslash(\) used to repressent  char that r difficult or impossible to type directly such as new line, tab , quote mark etc

```python
# new line - it moves the cursor to the next line: \n
print("Hello \nString")
# Tab - adds extra space bw words: \t
print("Name\tage")
print("Abhishek\t25")
# backslash - to print one backslash use 2 backslashes(\\)
print("C:\\Users\\Abhishek")
# single quote - to print a sinle quote inside a string
print("I\m learning AI ML") #it will not work
print('I\m learning AI ML')
```
