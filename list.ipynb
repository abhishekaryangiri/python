##### List
- List is an ordered,mutable(changable) collection in python that allows duplicate values and can store dufferent types of data. It support CRUD operation.

- positive index: start from 0.
- negative index: start from -1

```python
# CRUD(create, read, update, delete)
#  Create
rough_list = ["Mango", "Banana",100,10.5,True]


# Read
print(rough_list)
print(rough_list[-2])
print(type(rough_list))
# For access of all value use :
print(": ",rough_list[:])

# Update : use close bracket for indexing using nums+ve , -ve or 0.
rough_list[4] = "New Delhi"
print(rough_list)

# Delete
rough_list.remove(10.5)
rough_list.remove("Banana")
print(rough_list)

# Use pop() to remove last element
rough_list.pop()
print(rough_list)

```

---
- List slicing : retriving specific data/ a part from list.
seq[start : stop : step]

start: start index
stop: ending index
step: num of position to skip(optional)

```python
rough_list2 = ["Mango", "Banana",100,10.5,True]

# slicing - copy/return complete list  :: or :
print(rough_list2[::])

# basic slicing - print starting index to ending index list :
print(rough_list2[1:4])

# slice from beginning :number
list2 = ["Mango", "Banana",100,10.5,True]
print(list2[:2])
```

```python
# slice till end number:
list2 = ["Mango", "Banana",100,10.5,True]
print(list2[1:])
```

```python
list2 = ["Mango", "Banana",100,10.5,True]
list3 = ["Mango", "Banana",100,10.5,True]
list4 = ["Mango", "Banana",100,10.5,True]

# Negative index from end -num: or :num-
print(list2[-1:])

# -ve index from starting
print(list3[:-1])
```
