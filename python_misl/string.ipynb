## Escape Char:
#### that start with a backslash(\) used to perform special actions inside a string, suc as creating a new line, adding a tab , printig quotes and backslashes.

```python
# new line - it moves the cursor to the next line/ print end of the line: \n
print("Hello \nString")
# Tab - adds extra space bw words: \t
print("Name\tage")
print("Abhishek\t25")
# backslash - to print one backslash use 2 backslashes(\\)
print("C:\\Users\\Abhishek")
# single quote - to print a sinle quote inside a string
print("I\'m learning AI ML")
print('I\'m learning AI ML')
# double quote - use \" to print "(double quote) : \"\"
print("He said \"Hello\"")
# backspace - remove the previous char(\b)
print("ABC\bD")
# carriage return - moves the cursor to the beginning of the line., next text start writing from the beginning(\r)
print("Hell\roo")
print("abcd\refg")
# raw string - normally python reads escape chars like \n and \t,raw string tell python to treat them as normal text.(r"")
print(r"C:\Users\Abhishek\Documents")
```
