##### 1. List
Ordered and structured data store,  mutable coll., that allows duplicate val, store any typ of data, we can apply crud op.

```python

fruits = ["Mango","Apple","Banana","Guava"]
print(fruits)
print(type(fruits))
# Access element
print(fruits[0])
print(fruits[-1])
# Add items
fruits.append("Orange")
print(fruits)
# Used for SHopping cart, AI Dataset etc
```

```python
lst=list()
print(lst)
# Store diff data typ
a_list = ["Ram","Samsung","Str",100,500,True,False]
print(a_list)
# access data
print(a_list[6])
```

#####  2. Tuple
Ordered but immutable collection.

```python
colors = ("Red","Blue","Orange","Yellow")
print(colors)
print(type(colors))

#Immutable
#colors[0] = "Green"
#print(colors)
```

##### 3. Set
Unordered collection of unique values.

```python
nums = {10,20,30,40}
print(nums)
print(type(nums))
# Dup. not allowed
nums1 = {10,20,30,40,30}
print(nums1)
# add items
nums1.add(100)
print(nums1)
```

##### Dictionary
Store data as key-value pairs. Ex.
JSON data, API res.,Machine learning config.

```python
emp = {
    "name":"Abhishek",
    "company":"MoRD",
    "desg.":"Software Developer",
    "exp":"5+ yrs"
}
print(emp)
print(type(emp))

# Access val.
print(emp["company"])
# Add new Key
emp["location"] = "2nd Floor, MoRD,Jeevan Bharti Building, Cannaught Place"
print(emp)
```
